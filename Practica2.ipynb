#Random walk o paseo aleatorio
##En una dimensión:
Partimos de un valor inicial arbitrario, por ejemplo 1000, que podemos pensar que es la cotización de un bono en el mercado secundario.

En cada paso la cotización se incrementa o decrementa un cierto porcentaje con una probabilidad del 50% hacia arriba y otro 50% hacia abajo.

```python
import random
import matplotlib.pyplot as plt
random.seed()
cotiza = 1000     # inicializamos el valor de la cotización
n = 10000         # establecemos el número de sesiones cotizadas
factor = 1.01    # establecemos el factor. Si el incremento es del 1% el factor será 1.01
p = .5           # probabilidad de subida. Si es 0.5 la probabilidad de que el precio suba es del 50%
serie = [cotiza] # inicializamos la serie de precios, comenzamos con la primera cotización
for t in range(1,n):
    if random.random() < p:
        cotiza *= factor
    else:
        cotiza /= factor
    serie.append(cotiza)
plt.plot(range(n),serie)
plt.show()
```

##En dos dimensiones: 
Partimos del punto (x,y)=(0,0) origen del eje de coordenadas en el plano cartesiano.

Es el "paseo del borracho":

```python
# El caminar del borracho  
# Un paseo aleatorio en dos dimensiones  
# Se parte del punto (0,0) del plano cartesiano  
# Cada paso que se da es de longitud r  
# El ángulo con el que se da cada paso es aleatorio  
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=100 # número de pasos  
r=1 # longitud del paso  
x=y=0 # punto de partida (x,y)=(0,0)  
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()
```
