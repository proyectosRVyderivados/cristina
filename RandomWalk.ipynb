# Paseo aleatorio
* Partimos de un valor inicial arbitrario, por ejemplo 100, que podemos pensar que es la cotización de un valor bursátil
* En cada paso la cotización se incrementa o decrementa un cierto porcentaje con una probabilidad del 50% hacia arriba y otro 50% hacia abajo.

```python
import random
import matplotlib.pyplot as plt
random.seed()

cotiza = 100     # inicializamos el valor de la cotización
n = 1000         # establecemos el número de sesiones cotizadas
factor = 1.02    # establecemos el factor. Si el incremento es del 2% el factor será 1.02
p = .5           # probabilidad de subida. Si es 0.5 la probabilidad de que el precio suba es del 50%
serie = [cotiza] # inicializamos la serie de precios, comenzamos con la primera cotización
for t in range(1,n):
    if random.random() < p:
        cotiza *= factor
    else:
        cotiza /= factor
    serie.append(cotiza)
plt.plot(range(n),serie)
plt.show()
```

# Randoma Walk 2D
## Método 1
* Paseo de borracho.  
* Es un paseo aleatorio en dos dimensiones.
* Partimos del punto (x,y)=(0,0) origen del eje de coordenadas en el plano cartesiano.

```python
# El caminar del borracho  
# Un paseo aleatorio en dos dimensiones  
# Se parte del punto (0,0) del plano cartesiano  
# Cada paso que se da es de longitud r  
# El ángulo con el que se da cada paso es aleatorio  
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=100 # número de pasos  
r=1 # longitud del paso  
x=y=0 # punto de partida (x,y)=(0,0)  
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 # en radianes  
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()
```

## Método 2
* Creamos dos series independientes
* Son dos paseos aleatorios 2D que representamos con colores diferentes

```python
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=100000 # número de pasos  
r=1 # longitud del paso  
x1=y1=x2=y2=0 # punto de partida (x,y)=(0,0)  
lista1x=[x1]  
lista1y=[y1]  
lista2x=[x2]  
lista2y=[y2]  
for i in range(n):  
  angulo1=random.random()*math.pi*2 # en radianes  
  angulo2=random.random()*math.pi*2 # en radianes  
  x1+=r*math.cos(angulo1)  
  y1+=r*math.sin(angulo1)  
  x2+=r*math.cos(angulo2)  
  y2+=r*math.sin(angulo2)  
  lista1x.append(x1)  
  lista1y.append(y1)  
  lista2x.append(x2)  
  lista2y.append(y2)  
plt.plot(lista1x,lista1y)  
plt.plot(lista2x,lista2y,'--r')  
plt.show()
```
